# Classifying Products into Various Categories like Male,Female,Kids & except all these three Products is Other Products

```python
import pandas as pd

# Assuming you have an existing DataFrame called 'df'

# Create a function to classify products into categories
def classify_category(sub_title):
    if 'Men' in sub_title:
        return 'Male Products'
    elif 'Women' in sub_title:
        return 'Female Products'
    elif 'Kids' in sub_title:
        return 'Kids Products'
    else:
        return 'Other Products'

# Add a 'Category' column to the existing DataFrame based on 'sub_title'
df['Category'] = df['sub_title'].apply(classify_category)

# You can now access the different categories using DataFrame filtering
df_male_products = df[df['Category'] == 'Male Products']
df_female_products = df[df['Category'] == 'Female Products']
df_kids_products = df[df['Category'] == 'Kids Products']
df_other_products = df[df['Category'] == 'Other Products']
```

# Dividing Products into certain Products Types like Top and Bottom Wearing Items,Footwears,Crownwears(Cap,Hat,Gogles)Outerwears(Jacket,hoodies) and except these five types is the Other Items

```python
# Create a function to categorize products
def categorize_product(row):
    if "T-Shirt" in row['sub_title'] or "Polo" in row['sub_title']:
        return "Top-Wears"
    elif "Pant" in row['sub_title'] or "Jeans" in row['sub_title'] or "Shorts" in row['sub_title']:
        return "Bottom-Wears"
    elif "Shoes" in row['sub_title'] or "Boots" in row['sub_title']:
        return "Foot-Wears"
    elif "Hat" in row['sub_title'] or "Cap" in row['sub_title']:
        return "Crown-Wears"
    elif "Jacket" in row['sub_title'] or "Hoodie" in row['sub_title']:
        return "Outer-Wears"
    else:
        return "Other Items"

# Apply the categorization function to create a new 'Category' column
df['Type'] = df.apply(categorize_product, axis=1)

# Create separate DataFrames for each category
top_wears_df = df[df['Type'] == "Top-Wears"]
bottom_wears_df = df[df['Type'] == "Bottom-Wears"]
foot_wears_df = df[df['Type'] == "Foot-Wears"]
crown_wears_df = df[df['Type'] == "Crown-Wears"]
outer_wears_df = df[df['Type'] == "Outer-Wears"]
other_items_df = df[df['Type'] == "Other Items"]
```

# Showing the Availability Status of Repeated Products and displaying proportion of 'In-Stock' & 'Out-Stock' for each repeated Product.

```python
import pandas as pd
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import string

# Download NLTK stopwords if not already downloaded
nltk.download('stopwords')
nltk.download('punkt')

# Assuming you already have the 'df_combined' DataFrame

# Define a function to clean the text
def clean_text(text):
    # Tokenize the text into words
    words = word_tokenize(str(text))
    
    # Remove punctuation and convert to lowercase
    words = [word.lower() for word in words if word.isalpha()]
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]
    
    # Join the cleaned words back into a string
    cleaned_text = ' '.join(words)
    
    return cleaned_text

# Apply the clean_text function to the 'description' column
df_combined['cleaned_description'] = df_combined['description'].apply(clean_text)

# Display the DataFrame with the cleaned description
print(df_combined[['description', 'cleaned_description']])
```

```python
from textblob import TextBlob
import pandas as pd

# Assuming you have the 'df_combined' DataFrame with the 'cleaned_description' column

# Define a function to get sentiment polarity and subjectivity
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Apply the sentiment analysis function to the 'cleaned_description' column
df_combined['sentiment'] = df_combined['cleaned_description'].apply(get_sentiment)

# Display the DataFrame with sentiment analysis results
print(df_combined[['cleaned_description', 'sentiment']])
```
