```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, MaxAbsScaler
from sklearn.preprocessing import FunctionTransformer
```

```python
# Specify the path to your Excel file
csv_file_path = "D:\\3DPDAP\\BostonHousing.csv"

# Read the Excel file into a DataFrame
df = pd.read_csv(csv_file_path)

# Now you can work with the DataFrame, for example, printing the first few rows
df
```

# Here, The Scailing and Transformation is Applied on 'nox' variable which is Nitric Oxide Concentration (PP10M)

```python
# Assuming your 'nox' variable is a DataFrame or Series
nox_data = df['nox']
```

```python
nox_data
```

# Normalization

```python
# Normalization
normalized_nox = nox_data.values.reshape(-1, 1)
normalized_nox = MinMaxScaler().fit_transform(normalized_nox)
print(normalized_nox)
```

# Standardization

```python
# Standardization
standardized_nox = StandardScaler().fit_transform(nox_data.values.reshape(-1, 1))
print(standardized_nox)
```

# Robust Scaling

```python
# Robust Scaling
robust_scaled_nox = RobustScaler().fit_transform(nox_data.values.reshape(-1, 1))
print(robust_scaled_nox)
```

# Min-Max Scaling

```python
# Min-Max Scaling
minmax_scaled_nox = MinMaxScaler().fit_transform(nox_data.values.reshape(-1, 1))
print(minmax_scaled_nox)
```

# Max Absolute Scaling

```python
# Max Absolute Scaling
maxabs_scaled_nox = MaxAbsScaler().fit_transform(nox_data.values.reshape(-1, 1))
print(maxabs_scaled_nox)
```

# Log Transformation

```python
# Log Transformation
log_transformed_nox = FunctionTransformer(func=np.log1p, validate=True).fit_transform(nox_data.values.reshape(-1, 1))
print(log_transformed_nox)
```

# Square Root Transformation

```python
sqrt_transformed_nox = FunctionTransformer(func=np.sqrt, validate=True).fit_transform(nox_data.values.reshape(-1, 1))
print(sqrt_transformed_nox)
```

# Reciprocal Transformation

```python
# Reciprocal Transformation
reciprocal_transformed_nox = FunctionTransformer(func=np.reciprocal, validate=True).fit_transform(nox_data.values.reshape(-1, 1))
print(reciprocal_transformed_nox)
```

# Histograms for All Scaling and Transformation Method

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have the transformed variables like normalized_nox, standardized_nox, etc.

# Define the transformations and labels
transformations = [nox_data, normalized_nox, standardized_nox, robust_scaled_nox, minmax_scaled_nox, maxabs_scaled_nox, log_transformed_nox, sqrt_transformed_nox, reciprocal_transformed_nox]
labels = ['Original', 'Normalized', 'Standardized', 'Robust Scaled', 'Min-Max Scaled', 'Max Absolute Scaled', 'Log Transformed', 'Square Root Transformed', 'Reciprocal Transformed']

# Create a 3x3 grid of histograms
plt.figure(figsize=(15, 15))
for i in range(9):
    plt.subplot(3, 3, i + 1)
    sns.histplot(transformations[i], kde=True)
    plt.title(labels[i])

plt.tight_layout()
plt.show()
```
