```python
def clean_dataframe(df):
    # Remove leading and trailing whitespaces from all columns
    df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)

    # Replace or remove specific junk characters in all columns
    df = df.applymap(lambda x: x.replace('!', '') if isinstance(x, str) else x)

    return df

# Clean the entire DataFrame using the function
new_df = clean_dataframe(df)

# Print the cleaned DataFrame
new_df
```

Function for Removing White Spaces & Junk Characters from whole Dataframe

```python
# Function to clean a string column
def clean_string(s):
    if isinstance(s, str):
        # Remove leading and trailing white spaces
        s = s.strip()
        # Remove non-alphanumeric characters
        s = ''.join(e for e in s if e.isalnum() or e.isspace())
    return s

# Apply the cleaning function to all string columns
string_columns = new_df.select_dtypes(include='object')
new_df[string_columns.columns] = string_columns.applymap(clean_string)
```
